```python
def split_train_test(X,test_ratio=0.2):
    shuffled_indices = np.random.permutation(len(X))
    train_size=int(len(X)*(1-test_ratio))
    #train_size=len(X)*test_ratio
    train_indices=shuffled_indices[:train_size]
    test_indices=shuffled_indices[train_size:]
    
    return X.iloc[train_indices],X.iloc[test_indices]
```

### Custom Transformer
    *combines 3 columns

```python
housing = strat_train_set.copy()
housing.columns
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
```

#### Conditions for creating Transformeer
    *what kind of input?
        - design principles: user numpy arrays
    * what kind of output?
        - 2d array with increased number of columns

```python
class Combined_att_adder(BaseEstimator, TransformerMixin):
    def __init__(self):
        self.total_rooms_ix = 3
        self.total_bedrooms_ix = 4
        self.households_ix = 6
        
    def fit(self,X, y=None):                 #keep number of parameters in fit and transform same
        return self
    
    def transform(self, X, y = None):           #X is data, y is target 
        total_rooms_per_hhold = X[:, self.total_rooms_ix] / X[:, self.households_ix]
        total_bedrooms_per_hhold =  X[:, self.total_bedrooms_ix] / X[:, self.households_ix]
        
        return np.c_[X,total_rooms_per_hhold,total_bedrooms_per_hhold]                         
    #c_ to combine unknown number of columns c = column, _ = unkown
    # as we have included TranformerMixin we need not write fit_transform
```

```python
caa = Combined_att_adder()
```

```python
housing_combined = caa.fit_transform(housing.values)
```

```python
cols = list(housing.columns)
cols.extend(['total_rooms_per_hhold, total_bedrooms_per_hhold'])

housing_df = pd.DataFrame(housing_combined,columns = cols)
```

### Piplines
    * Nnumeric data
        *Applied Imputers
        *Combined Attributes
        *Scaling
    * Caterogrical Date
        * 1hot Encoder

```python
housing_num = strat_train_set.drop('ocean_proximity', axis = 1)
housing_cat = strat_train_set['ocean_proximity']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
num_pipeline = Pipeline([('imputer',SimpleImputer(strategy='median')),('adder',Combined_att_adder),('scalar',StandardScaler)])
#pipeline can be used as transformers
```

```python
housing_num_prepared = num_pipeline.fit_transform(housing_num)
```

```python
from sklearn.compose import ColumnTransformer
```

```python
full_pipeline = ColumnTransformer([('num_pipeline', num_pipeline,housing_num.columns), ('1hot', OneHotEncoder(),['ocean_proximity'])])
```
