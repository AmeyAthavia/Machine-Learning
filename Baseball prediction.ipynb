```python
ball_new = ball.dropna()
```

```python
ball_new
```

```python
ball_new['RD'] = ball_new['RS'] - ball_new['RA']
```

```python
ball_new
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = ball_new[['OBP','SLG']].values
Y = ball_new[['RS']].values
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(X,Y)
```

```python
print(lin_reg.intercept_)
print(lin_reg.coef_)
```

```python
#model for opponents obp and slg
x = ball_new[['OOBP','OSLG']].values
y = ball_new[['RA']].values
```

```python
lin_ra = LinearRegression()
lin_ra.fit(x,y)
```

```python
print(lin_ra.intercept_)
print(lin_ra.coef_)
```

```python
# predict winning from run difference
x = ball_new[['RD']].values
y = ball_new[['W']].values
```

```python
lin_w = LinearRegression()
```

```python
lin_w.fit(x,y)
```

```python
print(lin_w.intercept_)
print(lin_w.coef_)
```
